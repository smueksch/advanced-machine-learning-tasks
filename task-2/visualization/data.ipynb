# Visualization of Raw Signal

```python
import os
import sys
sys.path.append(os.path.join(os.pardir, os.pardir))

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from amlutils.task2.loading import load_train_set

from IPython.display import display
```

```python
path_to_data = os.path.join(os.pardir, 'data')
X_train, y_train = load_train_set(path_to_data)

display(X_train)
display(y_train)
```

```python
display(X_train.describe())
```

```python
# Plot hand-picked samples of each class.
num_rows = 4
num_cols = 1
fig, axs = plt.subplots(num_rows, num_cols, figsize=(12 * num_cols,6 * num_rows))

fig.set_facecolor('white')

data_row = 0
axs[0].plot(list(range(len(X_train.iloc[data_row]))), X_train.iloc[data_row])
axs[0].set(xlabel='Time', ylabel='Signal', title='Class 0 Sample')

data_row = 3
axs[1].plot(list(range(len(X_train.iloc[data_row]))), X_train.iloc[data_row])
axs[1].set(xlabel='Time', ylabel='Signal', title='Class 1 Sample')

data_row = 4
axs[2].plot(list(range(len(X_train.iloc[data_row]))), X_train.iloc[data_row])
axs[2].set(xlabel='Time', ylabel='Signal', title='Class 2 Sample')

data_row = 30
axs[3].plot(list(range(len(X_train.iloc[data_row]))), X_train.iloc[data_row])
axs[3].set(xlabel='Time', ylabel='Signal', title='Class 3 Sample')

plt.show()
```

```python
display(y_train.value_counts())
```

```python
# Plot hand-picked samples of each class with rolling average.
num_rows = 4
num_cols = 1
fig, axs = plt.subplots(num_rows, num_cols, figsize=(12 * num_cols, 8 * num_rows))

fig.set_facecolor('white')

opacity = 0.25
window = 100

data_row = 0
axs[0].plot(list(range(len(X_train.iloc[data_row]))), X_train.iloc[data_row], label='Raw Signal', alpha=opacity)
axs[0].plot(list(range(len(X_train.iloc[data_row]))), X_train.iloc[data_row].rolling(window=window).mean(), label='Moving Average')
axs[0].set(xlabel='Time', ylabel='Signal', title='Class 0 Sample')
axs[0].legend()

data_row = 3
axs[1].plot(list(range(len(X_train.iloc[data_row]))), X_train.iloc[data_row], label='Raw Signal', alpha=opacity)
axs[1].plot(list(range(len(X_train.iloc[data_row]))), X_train.iloc[data_row].rolling(window=window).mean(), label='Moving Average')
axs[1].set(xlabel='Time', ylabel='Signal', title='Class 1 Sample')
axs[1].legend()

data_row = 4
axs[2].plot(list(range(len(X_train.iloc[data_row]))), X_train.iloc[data_row], label='Raw Signal', alpha=opacity)
axs[2].plot(list(range(len(X_train.iloc[data_row]))), X_train.iloc[data_row].rolling(window=window).mean(), label='Moving Average')
axs[2].set(xlabel='Time', ylabel='Signal', title='Class 2 Sample')
axs[2].legend()

data_row = 30
axs[3].plot(list(range(len(X_train.iloc[data_row]))), X_train.iloc[data_row], label='Raw Signal', alpha=opacity)
axs[3].plot(list(range(len(X_train.iloc[data_row]))), X_train.iloc[data_row].rolling(window=window).mean(), label='Moving Average')
axs[3].set(xlabel='Time', ylabel='Signal', title='Class 3 Sample')
axs[3].legend(loc='upper right')

plt.show()
```

```python
# Plot class averages.
num_rows = 4
num_cols = 1
fig, axs = plt.subplots(num_rows, num_cols, figsize=(12 * num_cols,6 * num_rows))

fig.set_facecolor('white')

class_0_avg = X_train[y_train['y'] == 0].mean()
axs[0].plot(list(range(len(class_0_avg))), class_0_avg)
axs[0].set(xlabel='Time', ylabel='Signal', title='Class 0 Average')

class_1_avg = X_train[y_train['y'] == 1].mean()
axs[1].plot(list(range(len(class_1_avg))), class_1_avg)
axs[1].set(xlabel='Time', ylabel='Signal', title='Class 1 Average')

class_2_avg = X_train[y_train['y'] == 2].mean()
axs[2].plot(list(range(len(class_2_avg))), class_2_avg)
axs[2].set(xlabel='Time', ylabel='Signal', title='Class 2 Average')

class_3_avg = X_train[y_train['y'] == 3].mean()
axs[3].plot(list(range(len(class_3_avg))), class_3_avg)
axs[3].set(xlabel='Time', ylabel='Signal', title='Class 3 Average')

plt.show()
```

```python
# Plot class averages.
num_rows = 4
num_cols = 1
fig, axs = plt.subplots(num_rows, num_cols, figsize=(12 * num_cols,6 * num_rows))

fig.set_facecolor('white')

class_0_avg = X_train[y_train['y'] == 0].mean()
axs[0].plot(list(range(len(class_0_avg))), class_0_avg, alpha=0.5)
axs[0].set(xlabel='Time', ylabel='Signal', title='Class 0 Average')

class_1_avg = X_train[y_train['y'] == 1].mean()
axs[0].plot(list(range(len(class_1_avg))), class_1_avg, alpha=0.5)
axs[0].set(xlabel='Time', ylabel='Signal', title='Class 1 Average')

class_2_avg = X_train[y_train['y'] == 2].mean()
axs[0].plot(list(range(len(class_2_avg))), class_2_avg, alpha=0.5)
axs[0].set(xlabel='Time', ylabel='Signal', title='Class 2 Average')

class_3_avg = X_train[y_train['y'] == 3].mean()
axs[0].plot(list(range(len(class_3_avg))), class_3_avg, alpha=0.5)
axs[0].set(xlabel='Time', ylabel='Signal', title='Class 3 Average')

plt.show()
```

```python
# Plot class averages.
num_rows = 1
num_cols = 1
fig, axs = plt.subplots(num_rows, num_cols, figsize=(12 * num_cols,6 * num_rows))

fig.set_facecolor('white')

class_0_avg = X_train[y_train['y'] == 0].mean()
axs.plot(list(range(8000)), class_0_avg[:8000], alpha=0.5, label='Class 0')

class_1_avg = X_train[y_train['y'] == 1].mean()
axs.plot(list(range(8000)), class_1_avg[:8000], alpha=0.5, label='Class 1')

class_2_avg = X_train[y_train['y'] == 2].mean()
axs.plot(list(range(8000)), class_2_avg[:8000], alpha=0.5, label='Class 2')

class_3_avg = X_train[y_train['y'] == 3].mean()
axs.plot(list(range(8000)), class_3_avg[:8000], alpha=0.5, label='Class 3')

axs.set(xlabel='Time', ylabel='Signal', title='Average Singals per Class')
axs.legend()

plt.show()
```
